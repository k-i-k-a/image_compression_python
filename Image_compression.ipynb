##Given Helper Functions##

```python
def get_size(image):
  return image.size
```

```python
def quantization(image, n):
  return image.quantize(n)
```

**Modified Cell**

```python
def get_pixel_value(img,col,row):
  return img.getpixel((col,row))
```

```python
def get_unique_values(image):
  return np.unique(image), len(np.unique(image))
```

```python
def extract_row_color(arr,row,color):
  filtered = filter(lambda item: ((item[0] == row) and (item[-1] == color)), arr)
  return list(filtered)
```

```python
def merge_row_color(filtered, row, color):
  remove_row_color = [item[1:-1] for item in filtered]
  merge_runs = (row,)
  for item in remove_row_color:
    merge_runs = merge_runs + item
  otp = merge_runs + (color,)
  return otp
```

##Task 1##
**Task 1: Return the unique colors before and after quantization**

```python
def show_image_information(image):
  original_colors = []
  quantized_colors=[]
  # ......start code......
  (original_colors,x)= get_unique_values(image)
  quantized=quantization(image,9)
  (quantized_colors,y)=get_unique_values(quantized)

  # ......end code......
  return original_colors,quantized_colors
```

##Task 2##
**Task 2: Compute the runs per each row for the image**

```python
def compute_runs(image, unique_values):
  runs_code = []
  
  # ......start code......
  x,y= get_size(image)
  start=0
  for rows in range(y):
    for columns in range(x):
     if columns<x-1: 
        if get_pixel_value(image,columns,rows)==get_pixel_value(image,columns+1,rows):
          if(columns==x-2):
            runs_code=runs_code+[(rows,start,columns+1,get_pixel_value(image,columns,rows))]
          continue
        else:
          runs_code=runs_code+[(rows,start,columns,get_pixel_value(image,columns,rows))]
          start=columns+1
    else:
      if get_pixel_value(image,columns-1,rows)!=get_pixel_value(image,columns,rows):
        runs_code=runs_code+[(rows,start,columns,get_pixel_value(image,columns,rows))]
    start=0
  # ......end code......
  return runs_code
```

To test you code

```python
image_quantized = quantization(image, 9)
unique_values,length_unique_values = get_unique_values(image_quantized)
image_runs = compute_runs(image_quantized, unique_values)
image_runs
```

##Task 3##
**Compute the Run-Length Encoding of an Image**

```python
def compute_RLE(image, unique_values):
    result = []
    # ......start code......
    temp = compute_runs(image, unique_values)

    numberOfColors = len(unique_values)
    numberOfRows = get_size(image)[1]

    for y in range(numberOfRows):
      for x in range(numberOfColors):
        res=merge_row_color(extract_row_color(temp,y,unique_values[x]),y,unique_values[x])
        if(len(res)>2):
          result = result + [res]
    # ......end code......
    return result
```
